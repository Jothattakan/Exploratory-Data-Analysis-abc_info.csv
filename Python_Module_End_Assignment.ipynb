<h2 align="center"><font color="blue"> 1. Distribution of employees across each team and calculate the percentage split relative to the total number of employees.</font></h2

```python
team_counts = df['Team'].value_counts()
print(team_counts)
```

```python
total_employee = len(df)
print(total_employee)
```

```python
percentage_split = (team_counts / total_employee) * 100
percentage_split = percentage_split.round(2)
print(percentage_split)
```

```python
plt.figure(figsize = (10, 8))
plt.bar(team_counts.index, percentage_split)
plt.xlabel('Teams')
plt.ylabel('Percentage')
plt.title('Teams Vs Precentage spli of employee')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

<span style = 'background:yellow'>From this we come to know that the team “New Orleans Pelican”’  have the highest percentage split of employees with 4.15 % and the team “Orlando Magic” and “Minnesota Timberwolves “ share the lowest percentage split of employees with 3.06 %. Most of the teams have the percentage split of employees is 3.28%.</span>

<h2 align="center"><font color="blue">2. Segregate employees based on their positions within the company.</font></h2

```python
position = df['Position'].value_counts()
print(position)
```

```python
group_by_position = df.groupby('Position')['Name'].apply(list).reset_index()
print(group_by_position)
```

```python
plt.figure(figsize=(10, 8))
plt.bar(position.index, position.values, color = 'green')
plt.xlabel('Position')
plt.ylabel('Number of employees')
plt.title('Position Vs Number of employees')
plt.grid(True)
plt.show()
```

<span style = 'background:yellow'>From the above data we can conclude that the position "SG" contains the highest number of employees with 100 employees and the position "C" having the least number of employees.</span>

<h2 align="center"><font color="blue">4. Discover which team and position have the highest salary expenditure.</font></h2

```python
team_salary_expenditure = df.groupby('Team')['Salary'].sum()
print(team_salary_expenditure)
```

```python
plt.figure(figsize=(10, 8))
plt.pie(team_salary_expenditure, labels=team_salary_expenditure.index, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
position_salary_expenditure = df.groupby('Position')['Salary'].sum()
print(position_salary_expenditure)
```

```python
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(position_salary_expenditure, labels=position_salary_expenditure.index, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
highest_salary_team = team_salary_expenditure.idxmax()
print(highest_salary_team)
```

```python
highest_salary_position = position_salary_expenditure.idxmax()
print(highest_salary_position)
```

```python
plt.figure(figsize = (10, 8))
sns.boxplot(x = 'Team', y = 'Salary', data = df)
plt.title('Team Vs Salary')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.boxplot(x = 'Position', y = 'Salary', data = df)
plt.title('Position Vs Salary')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

<span style = 'background:yellow'>From the above boxplot data “Team Vs Salary”and “Position Vs Salary” we come to know that the highest salary expenditure team is “Cleveland Cavaliers“ with 106988689.00(5.00 %) and position is ”C” with 466377332.00(21.60 %).</span>
